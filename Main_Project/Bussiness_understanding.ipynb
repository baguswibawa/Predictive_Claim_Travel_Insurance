### <b><span style='color:coralred'> 1.5. Succsess Criteria</span></b>

**Target:**
- 0 : Tidak mengajukan klaim
- 1 : Mengajukan klaim

#### Miss-prediction Conditions and Consequences
**a. Classification:**

---

**i. Type 1 error (False Positive):**  
Predicted: Pemegang polis akan mengajukan klaim  
Actual: Pemegang polis tidak mengajukan klaim  

**Consequence:**  
Pada kasus false positive, perusahaan asuransi akan mengalokasikan sumber daya untuk memantau atau melakukan tindakan pencegahan terhadap pemegang polis yang sebenarnya tidak akan mengajukan klaim. Berdasarkan estimasi rata-rata biaya administrasi per nasabah sebesar $4.23, ini akan menjadi biaya yang tidak perlu.

Selain itu, ada potensi biaya kesempatan yang hilang karena alokasi sumber daya ini bisa digunakan untuk pelanggan lain yang benar-benar berisiko tinggi. Total biaya yang terbuang dalam kasus false positive dapat mencapai **$4.23** per pemegang polis, di mana perusahaan juga mungkin mengeluarkan tambahan sekitar **$50** untuk tindakan pencegahan seperti inspeksi tambahan atau pemantauan berkelanjutan. Maka, total FP Cost bisa menjadi **$54.23** per pemegang polis.

---

**ii. Type 2 error (False Negative):**  
Predicted: Pemegang polis tidak akan mengajukan klaim  
Actual: Pemegang polis mengajukan klaim  

**Consequence:**  
Pada kasus false negative, perusahaan asuransi gagal mendeteksi risiko dari pemegang polis yang pada akhirnya mengajukan klaim. Berdasarkan rata-rata klaim dan biaya administrasi yang dihitung, perusahaan berisiko kehilangan biaya klaim yang signifikan. Jika klaim rata-rata adalah $42.34 dengan estimasi rata-rata biaya administrasi $4.23, maka total kerugian perusahaan akan mencapai $46.57 per pemegang polis. Namun, jika kita mempertimbangkan klaim besar yang tidak terantisipasi, kerugian ini dapat dengan mudah meningkat ke angka seperti **$10,000** per pemegang polis yang mengajukan klaim.

---

#### Assumptions
- **Profil Pelanggan:** Pemegang polis yang paling berisiko biasanya memiliki profil dengan beberapa faktor risiko, seperti tujuan perjalanan yang lebih berisiko atau durasi perjalanan yang lebih lama.
- **Biaya Pencegahan:** Rata-rata perusahaan menghabiskan $50 per pelanggan untuk tindakan pencegahan tambahan jika diprediksi berisiko tinggi.
- **Biaya Klaim:** Rata-rata biaya klaim yang dibayarkan adalah $10,000.

Keberhasilan model akan diukur berdasarkan kemampuannya untuk meminimalkan baik false positive maupun false negative, dengan fokus utama pada pengurangan false negative karena dampaknya yang lebih besar terhadap kerugian finansial. Tujuan akhir dari proyek ini adalah mengembangkan alat prediktif yang dapat secara akurat mengidentifikasi pelanggan yang berpotensi mengajukan klaim, memungkinkan perusahaan asuransi untuk memfokuskan sumber daya dan layanan pada pelanggan yang tepat, sehingga meningkatkan efisiensi operasional dan profitabilitas.

### <b><span style='color:coralred'> 1.6. Project Limitation</span></b>

Meskipun proyek ini bertujuan untuk meningkatkan efisiensi operasional dan mengurangi kerugian finansial dengan memprediksi pemegang polis yang kemungkinan besar akan mengajukan klaim, terdapat beberapa keterbatasan yang perlu dipertimbangkan:

1. **Keterbatasan Data:** Data yang digunakan untuk melatih model berasal dari data historis yang mungkin tidak mencakup semua variabel yang dapat mempengaruhi klaim, seperti perubahan regulasi, kondisi ekonomi, atau tren pasar yang muncul. Misalnya, data mungkin tidak mencerminkan dampak dari regulasi baru yang berlaku setelah periode data dikumpulkan.

2. **Kualitas Data:** Model prediksi sangat bergantung pada kualitas data yang digunakan. Jika data yang digunakan memiliki kualitas rendah, seperti adanya data yang hilang atau tidak lengkap, akurasi model akan terpengaruh secara signifikan. Penting untuk memastikan bahwa data yang digunakan mencakup variabel-variabel penting dengan tingkat kelengkapan dan keakuratan yang tinggi.

3. **Akuntabilitas Model:** Meskipun model prediktif ini diharapkan dapat meningkatkan efisiensi, prediksi model tidak akan 100% akurat. Oleh karena itu, perusahaan tidak dapat sepenuhnya bergantung pada hasil prediksi ini tanpa melakukan pertimbangan tambahan, seperti review manual atau penggunaan model prediksi lainnya sebagai pelengkap.

4. **Cakupan Variabel:** Model ini hanya mempertimbangkan variabel-variabel tertentu seperti usia, jenis kelamin, dan durasi perjalanan. Faktor-faktor lain seperti kondisi kesehatan, perilaku perjalanan, atau alasan perjalanan tidak termasuk dalam model ini, yang dapat membatasi keakuratan prediksi.

5. **Penerapan Hasil Prediksi:** Meskipun model ini dapat membantu mengidentifikasi pelanggan berisiko tinggi, keputusan akhir tetap memerlukan intervensi stakeholder. Pertimbangan lebih lanjut seperti kebijakan perusahaan, kapasitas sumber daya, dan risiko reputasi juga perlu diperhitungkan dalam menerapkan hasil prediksi ini.

```python
# Memilih kolom yang bertipe object atau category
cat_col = df.select_dtypes(include=['object', 'category']).columns

# Inisialisasi dictionary untuk menyimpan hasil chi-square
chi2_results = {}

# Melakukan uji chi-square untuk setiap pasangan kolom kategorikal
for col1 in cat_col:
    for col2 in cat_col:
        if col1 != col2:
            contingency_table = pd.crosstab(df[col1], df[col2])
            chi2_stat, p_val, dof, ex = chi2_contingency(contingency_table)
            chi2_results[(col1, col2)] = {'Chi2 Stat': chi2_stat, 'P-Value': p_val}

# Menampilkan hasil uji chi-square
for key, value in chi2_results.items():
    print(f"{key}: Chi2 Stat = {value['Chi2 Stat']}, P-Value = {value['P-Value']}")
```
